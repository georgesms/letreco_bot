```python
import os
import unidecode
import unicodedata
import string
from collections import Counter
```

```python
rootdir = 'wiki/data/text'
```

```python
if True:
    word_count = Counter()
    explored_subdir = []
else:
    with open('word_count.pkl', 'rb') as p:
        word_count = pickle.load(p)
    with open('explored_subdir.pkl', 'rb') as p:
        explored_subdir = pickle.load(p)
```

```python
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        dirfile = os.path.join(subdir, file)
        print(".",end="")
        if dirfile not in explored_subdir:
            if file.startswith("wiki_"):
                with open(dirfile, "r") as f:
                    document = f.read()
                    word_count.update(filter(lambda x: len(x)==5, unidecode.unidecode(document).translate(str.maketrans('', '', string.punctuation)).replace("\n", " ").split(" ")))
                
            explored_subdir.append(dirfile)
            
```
